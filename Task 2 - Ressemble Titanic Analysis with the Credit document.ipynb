```python
# Married&University dataframe i.e. the client is married and has a college degree
MarriedUni_df = credit[(credit['MARRIAGE'] == 1) & (credit['EDUCATION']==2)]
# Add MarriedUni column
MarriedUni_df['MarriedUni'] = 'MarriedUni'
# Not alone data frame i.e. the passenger has either a sibling or a parent.
NotMarriedNoUni_df = credit[(credit['MARRIAGE'] != 1) | (credit['EDUCATION']!=2)]
NotMarriedNoUni_df['MarriedUni'] = 'NotMarriedNoUni'
```

```python
# Merge the above dataframes
comb = [MarriedUni_df, NotMarriedNoUni_df]
# Merge and sort by index
credit = pd.concat(comb).sort_index()
```
